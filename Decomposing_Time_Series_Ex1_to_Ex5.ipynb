```python
from statsmodels.tsa.seasonal import seasonal_decompose
import pandas as pd
import matplotlib.pyplot as plt

# Example 1: Google Trends, number of searches of the word "vacation", time series shows seasonality
df_vacation = pd.read_csv(r'C:\Users\ungdu\Downloads\lab6_datamining_20521214\df_vacation.csv', parse_dates=['Month'], index_col=0)
ts1 = df_vacation['Num_Search_Vacation']
```

```python
# Decompose the time series
result = seasonal_decompose(ts1, model='additive', period=12)
result.plot()
plt.show()
```

```python
# Print the components
print(result.trend)
print(result.seasonal)
print(result.resid)
print(result.observed)
```
